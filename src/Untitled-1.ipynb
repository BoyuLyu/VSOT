```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

def generate_ellipsoid(a, b, c, n_theta=50, n_phi=50):
    """
    Generate the (x, y, z) coordinates for an ellipsoid
    with semi-axes a, b, c.
    """
    # Parameter ranges
    theta = np.linspace(0, 2*np.pi, n_theta)
    phi   = np.linspace(0,     np.pi,  n_phi)
    
    # Create a 2D mesh (theta, phi)
    theta, phi = np.meshgrid(theta, phi)

    # Parametric equations for the ellipsoid
    x = a * np.sin(phi) * np.cos(theta)
    y = b * np.sin(phi) * np.sin(theta)
    z = c * np.cos(phi)

    return x, y, z

def plot_ellipsoid(ax, x, y, z, color='blue', alpha=0.3):
    """
    Plot a single ellipsoid (x, y, z) on a given Axes3D instance.
    """
    ax.plot_surface(x, y, z, 
                    rstride=1, cstride=1, 
                    color=color, alpha=alpha, 
                    edgecolor='none')

# --- Example usage ---

# Define several sets of ratios (a, b, c).
ratios = [
    (1, 1, 1),   # Sphere
    (1,2,0.5)

]

fig = plt.figure(figsize=(12, 6))

# Create one 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Generate and plot each ellipsoid with a different color
colors = ['red', 'blue', 'green', 'purple', 'orange']

for (a, b, c), col in zip(ratios, colors):
    x, y, z = generate_ellipsoid(a, b, c)
    plot_ellipsoid(ax, x, y, z, color=col, alpha=0.4)

# Make the aspect ratio of the axes equal
ax.set_box_aspect([1,1,1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Series of Ellipsoids with Different Axis Ratios')

plt.show()
```
